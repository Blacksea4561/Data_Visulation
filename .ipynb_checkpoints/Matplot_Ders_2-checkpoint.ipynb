# Plotting multiple sets of data - Çoklu data  ve legend verme

```python
age = np.arange(25, 44, 2.5)
salary = np.arange(40000, 76000, 5000)
salary_2 = [45372, 52345, 54148, 94256,16544, 45600, 96542, 98148]
```

## Legend lokasyon verme fontsize - grafik çizgi sitili

```python
plt.plot(age, salary, "b:",label = " Salary" )
plt.plot(age, salary_2, "r-", label = " Salary_2")

plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend() #  1 ve 2  ci satirdaki  labellari  axisise basar
plt.title("Salary By Age");
```

```python
plt.style.use("seaborn-whitegrid")
```

```python
plt.figure(figsize =(10, 5))
plt.plot(age, salary, "b",label = " Salary" )
plt.plot(age, salary_2, "r", label = " Salary_2")

plt.xlabel("Age")
plt.ylabel("Salary")


plt.legend(bbox_to_anchor =(0.5, 0., 0.5, 0.5))

plt.title("Salary By Age");
```

```python
plt.figure(figsize =(10, 5))
plt.plot(age, salary, "b",label = " Salary" )
plt.plot(age, salary_2, "r", label = " Salary_2")

plt.xlabel("Age")
plt.ylabel("Salary")


plt.legend(loc = 2 ) # legend a ince ayar  vermek  için  önrnek  "best"

plt.title("Salary By Age");
```

```python
plt.figure(figsize =(10, 5))
plt.plot(age, salary, "b", linewidth=4, label = " Salary" )
plt.plot(age, salary_2, color = "green", linestyle= ":", label = " Salary_2")

plt.xlabel("Age")
plt.ylabel("Salary")


plt.legend(loc = 0)

plt.title("Salary By Age");
```

```python
plt.figure(facecolor = "c",figsize = (10, 5))

ax = plt.axes()
ax.set_facecolor("xkcd:mint green")

plt.plot(age, salary)
plt.plot(age, salary_2)

plt.plot(age, salary, "b", linewidth=4, label = " Salary" )
plt.plot(age, salary_2, color = "green", linestyle= ":", label = " Salary_2")

plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend(bbox_to_anchor=(0,0.5,0.5,0.5))
plt.title("Salary by Age")
plt.show()
```

```python
plt.figure(facecolor = "red", figsize=(10, 5))
ax = plt.axes()
ax.set_facecolor("xkcd:mint green")
plt.plot(age, salary)
plt.plot(age, salary_2)
plt.plot(age, salary, "b-", linewidth=4, label = "SALARY")           # linewidth = Çizgi kalınlığı = lw
plt.plot(age, salary_2, linestyle= ":", color="red", label = "SALARY_2") #linestyle = Çzigi sitili = ls
plt.xlabel("Age")
plt.ylabel("Salary")
plt.legend(bbox_to_anchor=(0, 0.5, 1, 2))
plt.title("salary by Age")
plt.show()
```

##  Function Method  and Object Oriented Methods

### Function Method

```python
age = np.arange(25, 44, 2.5)
salary = np.arange(40000, 76000, 5000)
salary_2 = [45372, 52345, 54148, 94256,16544, 45600, 96542, 98148]
```

```python
plt.style.use("seaborn-darkgrid")

plt.title("Salary By Age")
plt.ylabel("Salary")
plt.xlabel("Age")
#plt.grid(color='b', linestyle='-', linewidth=0.2)#  gridleri renklendirmek için kullanılabiir
plt.plot(age, salary)
#plt.plot(age, salary_2);
```

## Object Oriented Methods(OOM)

```python
plt.style.use("seaborn-darkgrid")

plt.title("Salary By Age")
plt.ylabel("Salary")
plt.xlabel("Age")
#plt.grid(color='b', linestyle='-', linewidth=0.2)#  gridleri renklendirmek için kullanılabiir
plt.plot(age, salary)
#plt.plot(age, salary_2);
```

```python
fig, ax =plt.subplots()

ax.plot(age, salary, "r")
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_title("Salary By Ages")
```
