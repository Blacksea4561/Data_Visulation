## Değişkenlerimizi tanımlayalım

m değerini output a göre numpy kütüphanesi yardımı ile oluşturalım

```python
m= np.linspace(0,10,11)
m
```

```python
m = np.arange(0., 11)
m
```

```python
m
```

```python
c = 3* 10**8
```

```python
E = m* c**2
```

```python
E
```

## Oluşturduğumuz dokümanları plotlayalım

#### PLOTTING E=mc^2

```python
plt.plot(m, E, "blue", lw = 5)
plt.title(label = " E = mc^2")

plt.xlabel("M in g")
plt.ylabel("E in j")
plt.xlim(0, 10);
```

```python
plt.plot(m, E, "orange", lw= 3)
plt.title(label = "E = mc^2")

plt.xlabel(" M in g ")
plt.ylabel(" E in j ")
plt.xlim(0,10)
plt.show()
           
```

```python
plt.figure(figsize = (8, 5))

plt.polt(m,E, color = "orange", lw = 3.5)
```

```python
plt.figure(figsize = (8, 5))

plt.plot(m, E, color = "orange", lw=3);
plt.title("E = mc^2")
plt.xlabel("M in g")
plt.ylabel("E in J")
plt.xlim(0,10);
```

## Alttaki görseli oluştururken farklı bir değişken tanımlamayın. Bence eksenlerden birinin logaritmik hesaplanması gerekebilir:)

## Ayrıca axes de oluşan gridler bir style değil bence matplotlib library de gridleri araştırmanız gerekir

```python
plt.plot(m, E, color = "green", lw = 3)
plt.title("E = mc^2")
plt.xlabel("M in g")
plt.ylabel("E in J")
plt.xlim(0, 10)
plt.yscale("log")
plt.grid(which='both', axis = "y");
```

```python
x = np.arange(0, 11)
y = x ** 2 

plt.figure(figsize = (8, 6))

plt.xlabel(" M in g ")
plt.ylabel(" E in J ")
plt.plot(x, y, "green", lw=2)
plt.yscale("log")
plt.title(label = "E = mc^2")
plt.grid(which='minor',axis="y",color = "r")
plt.ylim((10**17),(10**18))
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8,6))
plt.yscale("log")
plt.xlim(0, 10)
ax.plot(m, E, color = "green", linewidth = 2);
ax.set_xlabel("M in g", fontsize = 8);
ax.set_ylabel("E in J", fontsize = 8);
ax.set_title("E = mc^2", fontsize = 9);
ax.tick_params(labelsize = 8.0);
ax.grid(axis = "y", which = "both")
```

## Bu görselleri alabilenler için alkışı borç bilirim

```python
fig, ax1 = plt.subplots(figsize=(12,8))

ax1.plot(labels, july16_2007, lw = 5, color ="blue")
ax1.set_ylabel("July16_2007",  fontsize =20, color ="blue")

ax2= ax1.twinx()
ax2.plot(labels, july16_2020, lw = 5, color ="red")
ax2.set_ylabel("july16_2020",  fontsize =20 , color ="red");
```

```python
fig, ax1 = plt.subplots(figsize=(12,8))
ax1.spines["left"].set_color("blue")
ax1.spines["left"].set_linewidth(5)
ax1.spines["right"].set_color("red")
ax1.spines["right"].set_linewidth(5)
ax1.plot(labels, july16_2007, lw=3, color="blue")
ax1.set_ylabel("July 16 2007", fontsize = 20, color = "blue")
ax2 = ax1.twinx()
ax2.plot(labels, july16_2020, lw=3, color = "red")
ax2.set_ylabel("July 16 2020", fontsize = 20, color = "red");
```
