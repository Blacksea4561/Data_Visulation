## 2: Plot 12 most dangerous cities

```python
city = kill.city.value_counts().head(13)
city
```

```python
yvalue = city.values
yvalue
```

```python
xvalue = city.index
xvalue
```

```python
fig, ax =plt.subplots(figsize =(10, 6))

sns.barplot( x = xvalue, y = yvalue)

ax.set_xticklabels(city.index, rotation =45, color ="blue")
```

```python
# with seaborn
```

```python
# with seaborn countplot
```

```python
fig, ax = plt.subplots(figsize= (7, 6))
ax = sns.countplot(x = "city", data = kill, order= xvalue)
ax.set_xticklabels(city.index, rotation = 45)
ax.set_title("Most Dangerous Cities", fontsize = 14, color = "b")
plt.tight_layout()
```

```python
#ok
```

## 3: Plot 20 most dangerous states

```python
kill.state.value_counts().head(20)
```

```python
kill_state = kill.state.value_counts().head(20)
x_index = kill.state.value_counts().head(20).index
y_value = kill.state.value_counts().head(20).index
```

```python
fig, ax = plt.subplots(figsize = (10, 6))

sns.barplot(x = x_index, y = y_value)
ax.set_xticklabels(kill_state.index, rotaton =45, color="b")
ax.set_title("Most  dangerous states", fontsize = 14, color = "b")

plt.tight_layout()
```
