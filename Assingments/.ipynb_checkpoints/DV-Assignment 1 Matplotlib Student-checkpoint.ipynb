## Kütüphaneleri import edin

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Değişkenlerimizi tanımlayalım

m değerini output a göre numpy kütüphanesi yardımı ile oluşturalım

```python
m = np.linspace(0, 10, 11)
```

```python
m
```

```python
c = 3* 10**8
```

```python
E = m* c**2
```

```python
E
```

## Oluşturduğumuz dokümanları plotlayalım

#### PLOTTING E=mc^2

```python
plt.plot(m, E, color = "blue", lw=3);
plt.title("E = mc^2")
plt.xlabel("M in g")
plt.ylabel("E in J")
plt.xlim(0,10);
```

```python
plt.figure(figsize =(8,5))

plt.plot(m,E,color="orange", linewidth=3.5)
plt.xlabel("M in g")
plt.ylabel("E in J")
plt.title(label="E=mc^2" )
plt.show()
```

## Alttaki görseli oluştururken farklı bir değişken tanımlamayın. Bence eksenlerden birinin logaritmik hesaplanması gerekebilir:)

## Ayrıca axes de oluşan gridler bir style değil bence matplotlib library de gridleri araştırmanız gerekir

```python
plt.plot(m, E, color = "green", lw = 3)

plt.title("E = mc^2")
plt.xlabel("M in g")
plt.ylabel("E in J")
plt.xlim(0, 10)
plt.yscale("log")
plt.grid(which='both', axis = "y");
```

```python
fig, ax = plt.subplots(figsize = (8,6))
plt.yscale("log")
plt.xlim(0, 10)
ax.plot(m, E, color = "green", linewidth = 2);
ax.set_xlabel("M in g", fontsize = 8);
ax.set_ylabel("E in J", fontsize = 8);
ax.set_title("E = mc^2", fontsize = 9);
ax.tick_params(labelsize = 8.0);
ax.grid(axis = "y", which = "both")
```

## Aşağıdaki değişkenlere göre ilgili plotlamaları yapın

```python
labels = ["1 Mo", "3 Mo", "6 Mo", "1 Yr", "2 Yr", "3 Yr", "5 Yr", "7 Yr", "10 Yr", "20 Yr", "30 Yr"]
july16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
july16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

```python
plt.figure(figsize = (8, 6))

plt.plot(labels, july16_2007, label = "July16 2007")
plt.plot(labels, july16_2020, label = "July16 2020")
plt.legend(loc = 5);
```

```python
plt.figure(figsize = (8, 6))

plt.plot(labels, july16_2007, label = "July16 2007")
plt.plot(labels, july16_2020, label = "July16 2020")
plt.legend(loc = (1.01, 0.4));
```

```python
plt.figure(figsize = (8, 6))

plt.plot(labels, july16_2007, lw = 3, label = "July16 2007")
plt.plot(labels, july16_2020, lw = 3, label = "July16 2020")
plt.legend(loc = (1.01, 0.4));
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 1, figsize = (12, 8))

ax[0].plot(labels, july16_2007, lw = 3, label = "July16 2007")
ax[0].set_title("July16 2007")

ax[1].plot(labels, july16_2020, lw = 3, label = "July16 2020")
ax[1].set_title("July16 2020");
```

## Alttaki görselleri araştırarak yapmaya çalışalım. Zihnimizi ve parmaklarımızı zorlayalım:)

#### Bir hatırlatma "fig, ax = plt.subplots" syntax i size cevabı bulmanızda yardımcı olur. 

#### ayrıca shift + tab+ tab yapmayı unutmayın ve içindeki parametreleri denemeyi unutmayın.

```python
fig, ax1 = plt.subplots(figsize=(12, 8))

ax1.plot(labels, july16_2007, lw = 3, color = "blue")
ax1.set_ylabel("July16 2007", fontsize = 20, color = "blue")

ax2= ax1.twinx()
ax2.plot(labels, july16_2020, lw = 3, color = "red")
ax2.set_ylabel("July16 2020", fontsize = 20, color = "red");
```

```python
fig, ax1 = plt.subplots(figsize=(12,8))

ax1.spines["left"].set_color("blue")
ax1.spines["left"].set_linewidth(5)

ax1.spines["right"].set_color("red")
ax1.spines["right"].set_linewidth(5)

ax1.plot(labels, july16_2007, lw=3, color="blue")
ax1.set_ylabel("July 16 2007", fontsize = 20, color = "blue")


ax2 = ax1.twinx()
ax2.plot(labels, july16_2020, lw=3, color = "red")
ax2.set_ylabel("July 16 2020", fontsize = 20, color = "red");
```

# sizden gelenler

```python
fig, ax1 = plt.subplots(figsize=(12,8))

ax1.spines["left"].set_color("blue")
ax1.spines["left"].set_linewidth(2)

ax1.tick_params(axis='y', colors = "blue", labelsize = 12, direction= "inout")

ax1.spines["right"].set_color("red")
ax1.spines["right"].set_linewidth(2)

ax1.plot(labels, july16_2007, lw=3, color="blue")
ax1.set_ylabel("July 16 2007", fontsize = 20, color = "blue")


ax2 = ax1.twinx()
ax2.plot(labels, july16_2020, lw=3, color = "red")
ax2.set_ylabel("July 16 2020", fontsize = 20, color = "red");
```
