```python
trace1 = go.Scatter(x = df2014.world_rank,
                   y =df2014.citations,
                   mode = "markers",
                   name = "2014",
                   marker = dict(color = "rgb(255, 128, 255)"),
                   text = df2014.university_name)
trace2 = go.Scatter(x = df2014.world_rank,
                   y =df2014.citations,
                   mode = "markers",
                   name = "2015",
                   marker = dict(color = "rgb(255, 128, 2)"),
                   text = df2014.university_name)
trace3 = go.Scatter(x = df2014.world_rank,
                   y =df2014.citations,
                   mode = "markers",
                   name = "2016",
                   marker = dict(color = "green"),
                   text = df2014.university_name)


data =[trace1, trace2, trace3]

layout = dict(title = "Ctation vs World Rank of Top Unv. with 2014, 2015, 2016",
             title_x =0.5,
             title_y =0.9,
             xaxis =dict (title = "Word rank"),
             yaxis =dict(title = "Citation"))

fig = dict(data =data , layout= layout)
iplot(fig)
```
